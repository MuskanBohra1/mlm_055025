# **General Descriptions of Data**



- **Data Source**: The dataset originates from a comprehensive record of import and export transactions, capturing various attributes associated with each transaction.

## Data Description




### Dataset Overview  

The dataset, sourced from Kaggle ([Imports-Exports Dataset](https://www.kaggle.com/datasets/chakilamvishwas/imports-exports-15000)), contains information on international trade transactions. It consists of **15,000 rows** and **16 columns**, where each row represents a unique transaction. The dataset provides a comprehensive view of imports and exports conducted between 2019 and 2024, with a majority of transactions occurring in 2021 and 2022.  

### Data Features  

**Columns**:  
1. **Transaction_ID**: A unique identifier assigned to each transaction.  
2. **Country**: The country participating in the transaction, either as the origin (export) or destination (import).  
3. **Product**: The type of product traded.  
4. **Import_Export**: Specifies whether the transaction is an import or export.  
5. **Quantity**: The quantity of the traded product.  
6. **Value**: The total value of the transaction in USD.  
7. **Date**: The date of the transaction, formatted as DD-MM-YYYY.  
8. **Category**: Product category, such as Machinery, Clothing, Electronics, Furniture, or Toys.  
9. **Port**: The port where the transaction was executed.  
10. **Customs_Code**: A unique customs identifier for the transaction.  
11. **Weight**: The weight of the traded goods in kilograms.  
12. **Shipping_Method**: The transportation method used for the transaction (e.g., Air or Sea).  
13. **Supplier**: The company supplying the goods.  
14. **Customer**: The recipient of the traded goods.  
15. **Invoice_Number**: A unique identifier for the transaction's invoice.  
16. **Payment_Terms**: The payment conditions agreed upon, such as Cash on Delivery, Prepaid, or Net terms (e.g., Net 30, Net 60).

```python
# Spearman and Kendall's Ranl correlation


# Create an empty DataFrame to store the results
correlation_results = pd.DataFrame(index=mb25_sdf_cat.columns,
                                    columns=['Spearman', 'Kendall'])

# Loop through each pair of columns in the DataFrame
for col1 in mb25_sdf_cat.columns:
    for col2 in mb25_sdf_cat.columns:
        if col1 != col2:  # Avoid calculating correlation with itself
            # Calculate Spearman's rank correlation
            spearman_corr, _ = spearmanr(mb25_sdf_cat[col1],
                                         mb25_sdf_cat[col2],
                                         nan_policy='omit')  # Handle missing values

            # Calculate Kendall's rank correlation
            kendall_corr, _ = kendalltau(mb25_sdf_cat[col1],
                                         mb25_sdf_cat[col2],
                                         nan_policy='omit')  # Handle missing values

            # Store the results in the DataFrame
            correlation_results.loc[col1, 'Spearman'] = spearman_corr
            correlation_results.loc[col1, 'Kendall'] = kendall_corr

# Display the results
print("Correlation Results:\n", correlation_results)
```

### Non Categorical Descreptive Statistics

```python
# Assuming mb25_sdf is your DataFrame with the relevant columns
mb25_sdf_cat = mb25_sdf[['Import_Export', 'Category', 'Shipping_Method', 'Payment_Terms']]

# Function to run Chi-square test for homogeneity
def chi_square_test_of_homogeneity(df, columns):
    results = {}
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            col1 = columns[i]
            col2 = columns[j]
            contingency_table = pd.crosstab(df[col1], df[col2])

            # Perform Chi-square test
            chi2, p, dof, expected = chi2_contingency(contingency_table)

            results[f"{col1} vs {col2}"] = {
                "Chi2": chi2,
                "p-value": p,
                "Degrees of Freedom": dof,
                "Expected Frequencies": expected
            }
    return results

# Example usage
chi_square_results = chi_square_test_of_homogeneity(mb25_sdf_cat, mb25_sdf_cat.columns)

# Print results
for comparison, stats in chi_square_results.items():
    print(f"\n{comparison}:")
    for key, value in stats.items():
        if key == "Expected Frequencies":
            print(f"{key}:\n{value}")
        else:
            print(f"{key}: {value}")
```

### Non Categorical Inferential Statistics

```python
import pandas as pd
import numpy as np
from scipy import stats

# Assuming 'mb25_sdf_noncat' is already defined as a DataFrame with non-categorical columns

alpha = 0.05  # Significance level

# Loop through each column in the DataFrame
for column in mb25_sdf_noncat.columns:
    data = mb25_sdf_noncat[column].dropna()  # Remove missing values for calculations

    print(f"Variable: {column}")

    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p = stats.shapiro(data)
    print(f"  Shapiro-Wilk Test: statistic={shapiro_stat:.3f}, p-value={shapiro_p:.3f}, ", end="")
    if shapiro_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Kolmogorov-Smirnov Test
    ks_stat, ks_p = stats.kstest(data, 'norm')
    print(f"  Kolmogorov-Smirnov Test: statistic={ks_stat:.3f}, p-value={ks_p:.3f}, ", end="")
    if ks_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Anderson-Darling Test
    anderson_result = stats.anderson(data, dist='norm')
    print(f"  Anderson-Darling Test: statistic={anderson_result.statistic:.3f}, ", end="")
    if anderson_result.statistic < anderson_result.critical_values[2]:  # Using critical value for alpha=0.05
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    # Jarque-Bera Test
    jb_stat, jb_p = stats.jarque_bera(data)
    print(f"  Jarque-Bera Test: statistic={jb_stat:.3f}, p-value={jb_p:.3f}, ", end="")
    if jb_p > alpha:
        print("Data is normally distributed (fail to reject H0)")
    else:
        print("Data is not normally distributed (reject H0)")

    print("-" * 20)
```
